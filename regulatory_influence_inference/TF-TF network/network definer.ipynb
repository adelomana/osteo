```python
regulons_up_file = home_dir + 'gd15/hi/research/osteo/results/regulons/regulons_up.csv'
regulons_down_file = home_dir + 'gd15/hi/research/osteo/results/regulons/regulons_down.csv'

interactions_output_file = home_dir + 'gd15/hi/research/osteo/results/TF-TF_network/interactions.csv'
regulon_descriptions_output_file = home_dir + 'gd15/hi/research/osteo/results/TF-TF_network/descriptions.csv'
```

# 1. read regulons

```python
up = pandas.read_csv(regulons_up_file, index_col=0, sep='\t')
print(up.shape)
up.head()
```

```python
down = pandas.read_csv(regulons_down_file, index_col=0, sep='\t')
print(down.shape)
down.head()
```

# 2. establish interactions

```python
# get full list of TFs
TFs_up = up['TF'].to_list() 
print(len(TFs_up))
TFs_down = down['TF'].to_list()
print(len(TFs_down))

all_TFs = list(set(TFs_up + TFs_down))
print(len(all_TFs))
```

```python
interactions = []; all_targets = []
for TF in all_TFs:
    print(TF)
    
    # up
    container_list = up[up['TF'] == TF]['Target genes'].to_list()
    if len(container_list) > 0:
        targets = container_list[0].split(', ')
        print('\t up {}'.format(targets))
        for target in targets:
            if target in all_TFs:
                print('\t {}'.format(target))
                interactions.append((TF, target))
                for i in range(len(targets)):
                    all_targets.append(targets[i])
    
    # down        
    container_list = down[down['TF'] == TF]['Target genes'].to_list()
    if len(container_list) > 0:
        targets = container_list[0].split(', ')
        print('\t down {}'.format(targets))
        for target in targets:
            if target in all_TFs:
                print('\t {}'.format(target))
                interactions.append((TF, target))
                for i in range(len(targets)):
                    all_targets.append(targets[i])

print(len(list(set(all_targets))))
print(interactions, len(interactions))
```

# 3. store results

```python
with open(interactions_output_file, 'w') as f:
    f.write('source\ttarget\n')
    for interaction in interactions:
        f.write('{}\t{}\n'.format(interaction[0], interaction[1]))
```

```python
with open(regulon_descriptions_output_file, 'w') as f:
    f.write('node\tregulation\tsize\n')
    
    for element in up.index: 
        node = up.loc[element, 'TF']
        regulation = 'up'
        size = up.loc[element, 'Regulon size']
        f.write('{}\t{}\t{}\n'.format(node, regulation, size))
        
    for element in down.index: 
        node = down.loc[element, 'TF']
        regulation = 'down'
        size = down.loc[element, 'Regulon size']
        f.write('{}\t{}\t{}\n'.format(node, regulation, size))
```
